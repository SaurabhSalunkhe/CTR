```python
import timeit
import sqlite3
import pandas as pd
from pandas.io import sql
import numpy as np


# creating a connection object that represent the database
conn = sqlite3.connect('../database.sqlite')

# representing the query string
query_train = """
select * from trainSearchStream limit 60000;
"""

# Read SQl query into a dataframe 
df = sql.read_sql(query_train, conn)

df.replace('', np.nan, inplace=True)

new_df_dropped=df.dropna()


j=0
k=100

for i in range(1,11):
    df_files=new_df_dropped[j:k]
    df_files.to_csv('newTrain%s.csv'%(i),index=False,header=None)
    j=j+100
    k=k+100
```
